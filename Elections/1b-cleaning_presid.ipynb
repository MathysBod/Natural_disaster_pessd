This code cleans the presidential electoral databases so as to keep only information related to the green party

- Identify and select relevant variables only
- Create variable indicating type of election
- Create variable indicating date of the vote
- Save new dataset

```python
import pandas as pd
import numpy as np
import os
```

```python
# Show all rows and columns when printing
pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', None) 
```

## Présidentielles 2007

```python
pres_2007 = pd.read_excel("raw/pres_2007_communes.xls", sheet_name="Tour 1")
pres_2007.head()
```

```python
pres_2007.shape
```

```python
#Let's describe "code nuance" columns
code_nuance_cols_2007 = [column for column in pres_2007.columns if column.startswith('Nom')]
print(pres_2007[code_nuance_cols_2007])
```

```python
pres_2007.columns
```

```python
pres_2007_eco = pres_2007[['Code du département', 'Libellé du département', 'Code de la commune',
       'Libellé de la commune', 'Inscrits', 'Abstentions', 'Votants', 'Blancs et nuls', 'Exprimés', 'Voix.5', '% Voix/Ins.5', '% Voix/Exp.5']]

pres_2007_eco.head(2)
```

```python
#Rename columns to match the city dataset
pres_2007_eco.rename(columns={'Libellé du département': 'department_name', 'Code du département': 'department_number', 'Libellé de la commune': 'city_code', 'Code de la commune': 'zip_code'}, inplace=True)
```

```python
#Change col names so they're unform and clear
pres_2007_eco.rename(columns={'Voix.5' : 'voix_eco', '% Voix/Ins.5' : '% Voix/Ins_eco', '% Voix/Exp.5' : '% Voix/Exp_eco'}, inplace=True)
```

```python
#Add a column to specify election type
pres_2007_eco['election_type'] = 'presid'
#Add a column indicating election date. 1er round of the 2007 presidential election was on April 22nd, 2007
pres_2007_eco['election_day'] = '20070422'
#Save new dataset to csv format. Default separator is ",", put ";" instead to avoid French/english number confusion
pres_2007_eco.to_csv(r"C:\Users\camil\OneDrive - GENES\Documents\ENSAE\Projet\data\elections\intermediate\presid_2007_inter.csv", sep=";", index=False)
```

# Présidentielles 2017

```python
pres_2017 = pd.read_excel("raw/pres_2017_communes.xls", sheet_name="Feuil1")
pres_2017.head()
```

```python
pres_2017.shape
```

```python
pres_2017.columns
```

```python
#Rename initial variable
pres_2017.rename(columns={'N°Panneau' : 'N°Panneau.0', 'Sexe' : 'Sexe.0', 'Nom' : 'Nom.0', 'Prénom' : 'Prénom.0', 'Voix' : 'Voix.0', '% Voix/Ins' : '% Voix/Ins.0', '% Voix/Exp' : '% Voix/Exp.0'}, inplace=True)
```

```python
#Rename columns to match the city dataset
pres_2017.rename(columns={'Libellé du département': 'department_name', 'Code du département': 'department_number', 'Libellé de la commune': 'city_code', 'Code de la commune': 'zip_code'}, inplace=True)
```

```python
#According to the explorations carried out earlier, the green candidate Hamon was present in Nom, Nom.1, ..., Nom.10. Let's convert data to long format to clean it up
```

```python
pres_2017_long = pd.wide_to_long(pres_2017, stubnames=['N°Panneau', 'Sexe', 'Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp'], i=['zip_code', 'department_number'], j='order', sep='.', suffix='\\d+')
```

```python
pres_2017_long.head(30)
```

```python
#On ne conserve que les lignes concernant Benoît Hamon, le candidat qu'avaient rallié les écologistes
pres_2017_eco = pres_2017_long[pres_2017_long['Nom'] == 'HAMON']
pres_2017_eco.head(10)
```

```python
pres_2017_eco.columns
```

```python
pres_2017_eco = pres_2017_eco.drop(columns=['% Abs/Ins', '% Blancs/Ins', '% Blancs/Vot', '% Exp/Ins', '% Exp/Vot',
       '% Nuls/Ins', '% Nuls/Vot', '% Vot/Ins', 'Prénom', 'Sexe', 'N°Panneau', 'Nom'])
```

```python
pres_2017_eco.columns
```

```python
pres_2017_eco.shape
```

```python
#Change col names so they're uniform and clear
pres_2017_eco.rename(columns={'Voix' : 'voix_eco', '% Voix/Ins' : '% Voix/Ins_eco', '% Voix/Exp' : '% Voix/Exp_eco'}, inplace=True)
```

```python
pres_2017_eco.head(10)
```

```python
#Add a column to specify election type
pres_2017_eco['election_type'] = 'presid'
```

```python
#Add a column indicating election date. 1er round of the 2017 presidential election was on 23 april 2017
pres_2017_eco['election_day'] = '20170423'
```

```python
#Save new dataset to csv format. Default separator is ",", put ";" instead to avoid French/english number confusion
pres_2017_eco.to_csv(r"C:\Users\camil\OneDrive - GENES\Documents\ENSAE\Projet\data\elections\intermediate\presid_2017_inter.csv", sep=";", index=True)
```
