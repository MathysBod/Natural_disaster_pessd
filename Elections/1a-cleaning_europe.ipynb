This code cleans the european electoral databases so as to keep only information related to the green party

- Identify and select relevant variables only
- Create variable indicating type of election
- Create variable indicating date of the vote
- Save new dataset

```python
import pandas as pd
import numpy as np
import os
```

# Européennes 2019

```python
europ_2019 = pd.read_excel("raw/europ_2019_communes.xls", sheet_name="Resultats-definitifs-par-commun")
europ_2019.head()
```

Les partis écologistes ne sont pas directement identifiés comme tels. Il va falloir les sélectionner :

- Urgence écologie : tête de liste "BOURG Dominique" dans la base
- Europe Écologie : tête de liste "JADOT Yannick" dans la base
- Décroissance 2019 : tête de liste "DELFEL Thérèse"
- (Envie d'Europe écologique et sociale) : discutable car y'a le PS dedans, on choisit de ne pas l'inclure

Comme les colonnes sont homogènes, i.e. 1 colonne par candidat, on peut seulement garder les colonnes d'intérêt :)

```python
#How many political parties in there ? 
cols_2019 = [column for column in europ_2019.columns if column.startswith('Nom Tête de Liste')]
europ_2019[cols_2019].head()
```

```python
#Urgence écologie correspond aux ".8", Europe écologie correspond aux ".29", Décroissance 2019 correspond aux ".16"
#Donc on va vouloir garder ces variables là + les variables standard en début de dataset
```

```python
#Rename columns to match the city dataset
europ_2019.rename(columns={'Libellé du département': 'department_name', 'Code du département': 'department_number', 'Libellé de la commune': 'city_code', 'Code de la commune': 'zip_code'}, inplace=True)
```

```python
europ_2019.columns
```

```python
col_eco = [col for col in europ_2019.columns if col.endswith(('.8', '.16', '.29'))]
```

```python
# Liste des variables à garder
conserv = ['department_number', 'department_name', 'zip_code', 'city_code', 'Inscrits', 'Abstentions', 'Votants', 'Blancs', 'Nuls', 'Exprimés']
conserv_eco = conserv + col_eco
conserv_eco
```

```python
europ_2019_eco = europ_2019[conserv_eco]
```

```python
europ_2019_eco.head()
```

```python
europ_2019_eco = europ_2019_eco.drop(columns=[col for col in europ_2019_eco.columns if col.startswith(('N°Liste', 'Libellé Abrégé Liste', 'Nom Tête de Liste'))])
```

```python
europ_2019_eco.head()
```

```python
#Create variable summing ecological ballots
europ_2019_eco['voix_eco'] = europ_2019_eco['Voix.8'] + europ_2019_eco['Voix.16'] + europ_2019_eco['Voix.29']
europ_2019_eco['% Voix/Ins_eco'] = europ_2019_eco['% Voix/Ins.8'] + europ_2019_eco['% Voix/Ins.16'] + europ_2019_eco['% Voix/Ins.29']
europ_2019_eco['% Voix/Exp_eco'] = europ_2019_eco['% Voix/Exp.8'] + europ_2019_eco['% Voix/Exp.16'] + europ_2019_eco['% Voix/Exp.29']
```

```python
#Clean useless variables
europ_2019_eco = europ_2019_eco.drop(columns=[col for col in europ_2019_eco.columns if col.startswith(('Libellé Etendu Liste.', 'Voix.', '% Voix/Ins.', '% Voix/Exp.'))])
```

```python
europ_2019_eco.head()
```

```python
#Add a column to specify election type
europ_2019_eco['election_type'] = 'europ'
```

```python
#Add a column indicating election date. Election day for 2019 european elections was 25 mai 2019
europ_2019_eco['election_day'] = '20190525'
```

```python
#Save new dataset to csv format. Default separator is ",", put ";" instead to avoid French/english number confusion
europ_2019_eco.to_csv(r"C:\Users\camil\OneDrive - GENES\Documents\ENSAE\Projet\data\elections\intermediate\europ_2019_inter.csv", sep=";", index=True)
```

# Européennes 2014

```python
europ_2014 = pd.read_excel("raw/europ_2014_communes.xlsx", sheet_name="Feuil1")
europ_2014.head()
```

```python
#Rename columns to match the city dataset
europ_2014.rename(columns={'Libellé du département': 'department_name', 'Code du département': 'department_number', 'Libellé de la commune': 'city_code', 'Code de la commune': 'zip_code'}, inplace=True)
```

```python
# Cette fois, on a les codes nuances, et on avait vu dans la phase exploratoire qu'il y a du LVEC dans Nuance Liste.3, Nuance Liste.5, Nuance Liste.6 et Nuance Liste.7
# On va supprimer les colonnes inutiles pour alléger la base, puis passer le dataset en long, puis trier les columns selon LVEC
```

```python
## Alléger les colonnes
col_eco = [col for col in europ_2014.columns if col.endswith(('.3', '.5', '.6', '.7'))]
col_eco
```

```python
# Liste des variables à garder
conserv = ['department_number', 'department_name', 'zip_code', 'city_code', 'Inscrits', 'Abstentions', 'Votants', 'Blancs', 'Nuls', 'Exprimés']
conserv_eco = conserv + col_eco
conserv_eco
```

```python
europ_2014_eco = europ_2014[conserv_eco]
```

```python
europ_2014_eco = europ_2014_eco.drop(columns=[col for col in europ_2014_eco.columns if col.startswith(('N°Liste', 'Libellé Abrégé Liste', 'Nom Tête de Liste'))])
```

```python
europ_2014_eco.head()
```

```python
#Passage en long
europ_2014_long = pd.wide_to_long(europ_2014_eco, stubnames=['Nuance Liste', 'Voix', '% Voix/Ins', '% Voix/Exp'], i=['zip_code', 'department_number'], j='order', sep='.', suffix='\\d+')
```

```python
europ_2014_long.head()
```

```python
#On ne conserve que les lignes concernant le candidat écolo, ie Nuance Liste == "LVEC"
europ_2014_eco = europ_2014_long[europ_2014_long['Nuance Liste'] == 'LVEC']
europ_2014_eco.head(10)
```

```python
#Create copy of dataset
europ_2014_eco = europ_2014_eco.copy()
```

```python
europ_2014_eco.rename(columns={'Voix' : 'voix_eco', '% Voix/Ins' : '% Voix/Ins_eco', '% Voix/Exp' : '% Voix/Exp_eco'}, inplace=True)
```

```python
europ_2014_eco.head()
```

```python
#Add a column to specify election type
europ_2014_eco['election_type'] = 'europ'
```

```python
#Add a column indicating election date. Election day for 2014 european elections was 25 mai 2014
europ_2014_eco['election_day'] = '20140525'
```

```python
europ_2014_eco = europ_2014_eco.drop(columns="Nuance Liste")
```

```python
europ_2014_eco.head()
```

```python
#Save new dataset to csv format. Default separator is ",", put ";" instead to avoid French/english number confusion
europ_2014_eco.to_csv(r"C:\Users\camil\OneDrive - GENES\Documents\ENSAE\Projet\data\elections\intermediate\europ_2014_inter.csv", sep=";", index=True)
```
